```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = [2,5]
epsilon = np.linspace(0,10,1000)
kT = 0.1


fig,ax = plt.subplots()

for m in mu:
    ax.plot(epsilon,1/(np.exp((epsilon-m)/kT)+1))
   

ax.fill_between(epsilon,1/(np.exp((epsilon-mu[0])/kT)+1),1/(np.exp((epsilon-mu[1])/kT)+1),hatch=',',alpha=0.1)
plt.xticks([2,5],[r"$\mu_2$","$\mu_1$"])
plt.xlabel(r"$\epsilon$")
plt.ylabel("$\overline{n}$")
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.title(r"$\int dE (f_1(E) - f_2(E))$")
plt.savefig("/home/remy13127/Documents/ghpages/content/images/fermi_difference.png")
plt.show()
plt.close()
```

```python
mu2 = [2,2.1]
epsilon = np.linspace(0,10,1000)
kT = 0.5

fig,ax = plt.subplots()

for m in mu2:
    ax.plot(epsilon,1/(np.exp((epsilon-m)/kT)+1))

plt.xticks([2.05],[r"$\mu_2 \approx \mu_1$"])
plt.xlabel(r"$\epsilon$")
plt.ylabel("$\overline{n}$")
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.title(r"$k\Theta >> |\mu_1 - \mu_2|$")
plt.savefig("/home/remy13127/Documents/ghpages/content/images/fermi_highT.png")
plt.show()
plt.close()

mu = 2.0
epsilon = np.linspace(0,10,1000)
kT = 0.02

fig,ax = plt.subplots()

ax.plot(epsilon,1/(np.exp((epsilon-mu)/kT)+1),label='f')
ax.plot(epsilon,(1/(np.exp((epsilon-mu)/kT)+1))**2,':',label=r'$f^2$',c='red')
ax.plot(epsilon,1-1/(np.exp((epsilon-mu)/kT)+1),'--',label='(1-f)')
plt.xticks([2.0],[r"$\mu$"])
plt.xlabel(r"$\epsilon$")
plt.ylabel("$\overline{n}$")
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.legend()
plt.title(r"$k\Theta << |\mu_1 - \mu_2|$")
plt.savefig("/home/remy13127/Documents/ghpages/content/images/fermi_lowT.png")
plt.show()
plt.close()

```
