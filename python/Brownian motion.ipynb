```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi=np.pi
N=1000
D=0.5

x = np.linspace(-5,5,1000)
t = [0.1,0.5,2,5]

fig, ax = plt.subplots()

for time in t:
    ax.plot(x,N/np.sqrt(4*pi*D*time)*np.exp(-x**2/(4*D*time)),label="t = "+str(time)+' s')

#plt.ylabel(r"f(x,t)")	
plt.xlabel(r"x")
plt.title("Diffusion of $N$ = "+str(N)+" Brownian particles, with $D$ = "+str(D)+' cm.s')
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.yticks([])
ax.legend()
plt.savefig("/home/remy13127/Documents/ghpages/content/images/brownian_gaussian.png")
plt.show()
plt.close()
```

```python
import scipy.special

x = np.linspace(0.,3,100)
n = np.linspace(0,3,4)
def factorial_stirling(n):
    return(np.sqrt(2*pi*n)*(n/np.exp(1))**n)

fig,ax = plt.subplots()
ax.plot(x,factorial_stirling(x),label="$(2\pi n)^{1/2}n^n e^{-n}$")
ax.scatter(n,scipy.special.factorial(n),label="$n!$")
ax.plot(x,scipy.special.gamma(x+1),label="$\Gamma (n+1)$")
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.title("Stirling's approximation")
plt.legend()
plt.savefig("/home/remy13127/Documents/ghpages/content/images/stirling_approximation.png")
plt.show()
plt.close()
```
